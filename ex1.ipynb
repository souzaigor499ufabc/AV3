```python
from math import sqrt, sin, pi
import numpy as np
import matplotlib.pyplot as plt

# Wave parameters
wavelength = 5.0  # cm
k = 2*pi/wavelength
xi0 = 1.0  # amplitude in cm
separation = 20.0  # separation of centers in cm

# Grid parameters
side = 100.0  # side length of square in cm
points = 500  # number of grid points per side
spacing = side/points

# Centers of the two wave sources
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# Create grid and calculate interference pattern
xi = np.empty([points, points], float)
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2 + (y-y1)**2)
        r2 = sqrt((x-x2)**2 + (y-y2)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2)

# Create density plot
plt.figure(figsize=(8,8))
plt.imshow(xi, origin="lower", extent=[0,side,0,side])
plt.colorbar(label='Wave height (cm)')
plt.title('Wave Interference Pattern')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.gray()
plt.show()
```
