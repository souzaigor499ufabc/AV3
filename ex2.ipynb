```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def create_lattice():
    # Lattice parameters
    L = 5  # Size of lattice
    R = 0.3  # Radius of spheres
    
    # Create figure
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    # Lists to store coordinates and colors
    x_coords, y_coords, z_coords = [], [], []
    distances = []
    
    # Generate lattice points and calculate distances
    for i in range(-L, L+1):
        for j in range(-L, L+1):
            for k in range(-L, L+1):
                x_coords.append(i)
                y_coords.append(j)
                z_coords.append(k)
                # Calculate distance from center
                distance = np.sqrt(i**2 + j**2 + k**2)
                distances.append(distance)
    
    # Create scatter plot with color mapping
    scatter = ax.scatter(x_coords, y_coords, z_coords,
                        c=distances,
                        cmap='viridis',
                        alpha=0.6,
                        s=200)
    
    # Add colorbar
    plt.colorbar(scatter, label='Distance from center')
    
    # Configure plot
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Atomic Cubic Lattice\nColored by distance from center')
    
    # Set axis limits
    ax.set_xlim(-L-1, L+1)
    ax.set_ylim(-L-1, L+1)
    ax.set_zlim(-L-1, L+1)
    
    plt.show()

# Create visualization
create_lattice()
```
